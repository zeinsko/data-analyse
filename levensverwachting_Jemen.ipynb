```python
import pandas as pd
import matplotlib.pyplot as plt


```

```python
df = pd.read_csv('API_SP.DYN.LE00.IN_DS2_en_csv_v2_2579.csv', skiprows=4)
yemen = df[df['Country Name'] == 'Yemen, Rep.']

```

```python
pop_data = yemen.loc[:, '1960':'2023'].T
pop_data.columns = ['Population']  # dit is de naam nu

pop_data.index.name = 'Year'
pop_data.reset_index(inplace=True)
pop_data['Year'] = pop_data['Year'].astype(int)
```

```python
plt.figure(figsize=(12,6))
plt.plot(pop_data['Year'], pop_data['Population'])  # niet 'LifeExpectancy'
plt.title('Levensverwachting in Jemen (1960-2023)')
plt.xlabel('Jaar')
plt.ylabel('Bevolking (miljoenen)')
plt.grid(True)
plt.show()
```

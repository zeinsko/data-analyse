```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_2590.csv', skiprows=4)
yemen = df[df['Country Name'] == 'Yemen, Rep.']
```

```python
pop_data = yemen.loc[:, '1960':'2023'].T
pop_data.columns = ['Population']
pop_data.index.name = 'Year'
pop_data.reset_index(inplace=True)
pop_data['Year'] = pop_data['Year'].astype(int)
```

```python
plt.figure(figsize=(12,6))
plt.plot(pop_data['Year'], pop_data['Population'] / 1e6)
plt.title('Bevolkingsgroei in Jemen (1960-2023)')
plt.xlabel('Jaar')
plt.ylabel('Bevolking (miljoenen)')
plt.grid(True)
plt.show()
```
